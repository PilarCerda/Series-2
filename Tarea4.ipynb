### Se convierten a Serie de tiempo, las 7 AFPs

```r
Cuprum_ts<-ts(AFP$Cuprum, start = c(1990,1), frequency = 12)
Habitat_ts<-ts(AFP$Habitat, start = c(1990,1), frequency = 12)
Magistere_ts<-ts(AFP$Magistere, start = c(1990,1), frequency = 12)
PlanVital_ts<-ts(AFP$Plan.Vital, start = c(1990,1), frequency = 12)
Provida_ts<-ts(AFP$Provida, start = c(1990,1), frequency = 12)
StaMaria_ts<-ts(AFP$Sta.Maria, start = c(1990,1), frequency = 12)
Summa_ts<-ts(AFP$Summa, start = c(1990,1), frequency = 12)
```

### Seleccion del modelo Arima, adecuado a cada serie

```r
Arima_Cuprum <- auto.arima(Cuprum_ts, seasonal = FALSE)
Arima_Cuprum 
Arima_Habitat <- auto.arima(Habitat_ts, seasonal = FALSE)
Arima_Habitat 
Arima_Magistere <- auto.arima(Magistere_ts, seasonal = FALSE)
Arima_Magistere 
Arima_PlanVital <- auto.arima(PlanVital_ts, seasonal = FALSE)
Arima_PlanVital
Arima_Provida <- auto.arima(Provida_ts, seasonal = FALSE)
Arima_Provida
Arima_StaMaria <- auto.arima(StaMaria_ts, seasonal = FALSE)
Arima_StaMaria 
Arima_Summa <- auto.arima(Summa_ts, seasonal = FALSE)
Arima_Summa 
```

#### Indice de comovimiento para cada par de series de tiempo.

```r
#Comptuting the Codispersion Coefficient
coords <- cbind(1:170, rep(1,170))
z_C_H<-codisp(AFP$Cuprum, AFP$Habitat, coords)
z_C_M<-codisp(AFP$Cuprum, AFP$Magistere, coords)
z_C_PL<-codisp(AFP$Cuprum, AFP$Plan.Vital, coords)
z_C_PV<-codisp(AFP$Cuprum, AFP$Provida, coords)
z_C_SM<-codisp(AFP$Cuprum, AFP$Sta.Maria, coords)
z_C_S<-codisp(AFP$Cuprum, AFP$Summa, coords)


z_H_M<-codisp(AFP$Habitat, AFP$Magistere, coords)
z_H_PL<-codisp(AFP$Habitat, AFP$Plan.Vital, coords)
z_H_PV<-codisp(AFP$Habitat, AFP$Provida, coords)
z_H_SM<-codisp(AFP$Habitat, AFP$Sta.Maria, coords)
z_H_S<-codisp(AFP$Habitat, AFP$Summa, coords)


z_M_PL<-codisp(AFP$Magistere, AFP$Plan.Vital, coords)
z_M_PV<-codisp(AFP$Magistere, AFP$Provida, coords)
z_M_SM<-codisp(AFP$Magistere, AFP$Sta.Maria, coords)
z_M_S<-codisp(AFP$Magistere, AFP$Summa, coords)


z_PL_PV<-codisp(AFP$Plan.Vital, AFP$Provida, coords)
z_PL_SM<-codisp(AFP$Plan.Vital, AFP$Sta.Maria, coords)
z_PL_S<-codisp(AFP$Plan.Vital, AFP$Summa, coords)


z_PV_SM<-codisp(AFP$Provida, AFP$Sta.Maria, coords)
z_PV_S<-codisp(AFP$Provida, AFP$Summa, coords)

z_SM_S<-codisp(AFP$Sta.Maria, AFP$Summa, coords)

```

```r
# plotting codispersion and cross-correlation functions
par(mfrow = c(1,2))


#Cuprum
ccf(AFP$Cuprum, AFP$Habitat, ylab = "cross-correlation", max.lag = 20)
plot(z_C_H)
ccf(AFP$Cuprum, AFP$Magistere, ylab = "cross-correlation", max.lag = 20)
plot(z_C_M)
ccf(AFP$Cuprum, AFP$Plan.Vital, ylab = "cross-correlation", max.lag = 20)
plot(z_C_PL)
ccf(AFP$Cuprum, AFP$Provida, ylab = "cross-correlation", max.lag = 20)
plot(z_C_PV)
ccf(AFP$Cuprum, AFP$Sta.Maria, ylab = "cross-correlation", max.lag = 20)
plot(z_C_SM)
ccf(AFP$Cuprum, AFP$Summa, ylab = "cross-correlation", max.lag = 20)
plot(z_C_S)

#Habitat
ccf(AFP$Habitat, AFP$Magistere, ylab = "cross-correlation", max.lag = 20)
plot(z_H_M)
ccf(AFP$Habitat,AFP$Plan.Vital, ylab = "cross-correlation", max.lag = 20)
plot(z_H_PL)
ccf(AFP$Habitat, AFP$Provida, ylab = "cross-correlation", max.lag = 20)
plot(z_H_PV)
ccf(AFP$Habitat, AFP$Sta.Maria, ylab = "cross-correlation", max.lag = 20)
plot(z_H_SM)
ccf(AFP$Habitat, AFP$Summa, ylab = "cross-correlation", max.lag = 20)
plot(z_H_S)

#Magistere
ccf(AFP$Magistere, AFP$Plan.Vital, ylab = "cross-correlation", max.lag = 20)
plot(z_M_PL)
ccf(AFP$Magistere, AFP$Provida, ylab = "cross-correlation", max.lag = 20)
plot(z_M_PV)
ccf(AFP$Magistere, AFP$Sta.Maria, ylab = "cross-correlation", max.lag = 20)
plot(z_M_SM)
ccf(AFP$Magistere, AFP$Summa, ylab = "cross-correlation", max.lag = 20)
plot(z_M_S)

#PlanVital
ccf(AFP$Plan.Vital, AFP$Provida, ylab = "cross-correlation", max.lag = 20)
plot(z_PL_PV)
ccf(AFP$Plan.Vital, AFP$Sta.Maria, ylab = "cross-correlation", max.lag = 20)
plot(z_PL_SM)
ccf(AFP$Plan.Vital, AFP$Summa, ylab = "cross-correlation", max.lag = 20)
plot(z_PL_S)

#ProVida
ccf(AFP$Provida, AFP$Sta.Maria, ylab = "cross-correlation", max.lag = 20)
plot(z_PV_SM)
ccf(AFP$Provida, AFP$Summa, ylab = "cross-correlation", max.lag = 20)
plot(z_PV_S)

#ProVida
ccf(AFP$Sta.Maria, AFP$Summa, ylab = "cross-correlation", max.lag = 20)
plot(z_SM_S)

```

## Estimacion de d

```r
malla2=c(1:20)
AICD=c()


for (i in malla2){
AICD=c(AICD,tar(flu, 11, 8, i, is.constant1 = TRUE, is.constant2 = TRUE, transform = "no",
 center = FALSE, standard = FALSE, estimate.thd = FALSE, threshold=0.28, 
method = c("MAIC", "CLS")[1],  order.select = TRUE, print = FALSE)$AIC)}


AIC_D<-data.frame(malla2,AICD)
AIC_D
```

```r
n=length(flu)
AICD_nd=c()

for (i in malla2){
    nd=max(11,i)
    temp=AICD[i]/(n-nd)
    AICD_nd=c(AICD_nd,temp)}


df_AICD_nd<-data.frame(AIC_D,AICD_nd)
min(df_AICD_nd$AICD_nd)
df_AICD_nd
```
